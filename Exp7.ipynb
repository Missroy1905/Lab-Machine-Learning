```python
from sklearn.cluster import KMeans 
from sklearn.mixture import GaussianMixture
import sklearn.metrics as metrics 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
names = ['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']
dataset = pd.read_csv('Iris.csv', header=0, names= names)
```

```python
X = dataset.iloc[:, :-1]
label = {'Iris-setosa' : 0, 'Iris-versicolor' : 1, 'Iris-virginica' : 2}
y = dataset.iloc[:, -1].map(label)
colormap = np.array(['red' , 'lime', 'black'])
plt.figure(figsize=(14, 4))
plt.subplot(1,3,1)
plt.title('Real')
plt.scatter(X.PetalLengthCm, X.PetalWidthCm, c=colormap[y])
```

```python
# KMeans clustering 
model = KMeans(n_clusters=3, random_state=0).fit(X)
plt.figure(figsize=(14, 4))
plt.subplot(1,3,2)
plt.title('K Means')
plt.scatter(X.PetalLengthCm, X.PetalWidthCm, c=colormap[model.labels_])
print('Accuracy score of K Means :', metrics.accuracy_score(y, model.labels_))
```

```python
#GMM gaussian mixture model / EM algo 
gmm = GaussianMixture(n_components=3, random_state=0).fit(X)
y_cluster_gmm = gmm.predict(X)
plt.figure(figsize=(14, 4))
plt.subplot(1,3,3)
plt.title('GMM classification')
plt.scatter(X.PetalLengthCm, X.PetalWidthCm, c=colormap[y_cluster_gmm])
```
